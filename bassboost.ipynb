```python
import json
import subprocess
import sys

STAGES = {
    "01_categorical": [
        "--disable-aligned-lags",
        "--disable-route-priors",
        "--disable-office-features",
        "--disable-share-features",
        "--disable-per-horizon-alpha",
        "--disable-blend",
    ],
    "02_aligned_lags": [
        "--disable-route-priors",
        "--disable-office-features",
        "--disable-share-features",
        "--disable-per-horizon-alpha",
        "--disable-blend",
    ],
    "03_route_priors": [
        "--disable-office-features",
        "--disable-share-features",
        "--disable-per-horizon-alpha",
        "--disable-blend",
    ],
    "04_office_features": [
        "--disable-share-features",
        "--disable-per-horizon-alpha",
        "--disable-blend",
    ],
    "05_share_features": [
        "--disable-per-horizon-alpha",
        "--disable-blend",
    ],
    "06_per_horizon_alpha": [
        "--disable-blend",
    ],
    "07_blend": [],
}

STAGE_SEQUENCE = list(STAGES)
# По умолчанию запускаем только финальную конфигурацию.
# Если хочешь пройти всю лестницу экспериментов, замени на: STAGE_SEQUENCE
EXPERIMENTS_TO_RUN = STAGE_SEQUENCE[-1:]

COMMON_ARGS = [
    "--train-path", str(TRAIN_PATH),
    "--test-path", str(TEST_PATH),
    "--alpha-min", "0.90",
    "--alpha-max", "1.10",
    "--alpha-step", "0.001",
    "--blend-weight-step", "0.05",
]


def run_stage(stage_name: str) -> dict:
    model_dir = WORKDIR / "models" / stage_name
    output_dir = WORKDIR / "outputs" / stage_name
    submission_path = output_dir / "submission_from_train.csv"

    cmd = [
        sys.executable,
        "train.py",
        "--models-dir", str(model_dir),
        "--outputs-dir", str(output_dir),
        "--submission-path", str(submission_path),
        *COMMON_ARGS,
        *STAGES[stage_name],
    ]

    print(f"\n===== Running {stage_name} =====")
    print(" ".join(cmd))
    subprocess.run(cmd, check=True, cwd=WORKDIR)

    with (output_dir / "validation_scores.json").open("r", encoding="utf-8") as fp:
        metrics = json.load(fp)

    print("Final overall metrics:", metrics["overall"]["final"])
    return {
        "stage": stage_name,
        "model_dir": model_dir,
        "output_dir": output_dir,
        "submission_from_train": submission_path,
        "metrics": metrics,
    }


results = [run_stage(stage_name) for stage_name in EXPERIMENTS_TO_RUN]
results
```

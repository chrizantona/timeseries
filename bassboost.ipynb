```python
from pathlib import Path
import shutil

DATASET_DIR = Path("/kaggle/input/datasets/alanchoo/123123/kaggle_add_data_bundle")
WORKDIR = Path("/kaggle/working/baseline_target_2h")
WORKDIR.mkdir(parents=True, exist_ok=True)

for name in [
    "train.py",
    "infer.py",
    "requirements.txt",
    "train_team_track.parquet",
    "test_team_track.parquet",
]:
    shutil.copy2(DATASET_DIR / name, WORKDIR / name)

shutil.copytree(DATASET_DIR / "src", WORKDIR / "src", dirs_exist_ok=True)

print("DATASET_DIR =", DATASET_DIR)
print("WORKDIR =", WORKDIR)
print(sorted([p.name for p in WORKDIR.iterdir()]))
```
